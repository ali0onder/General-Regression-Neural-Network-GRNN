```python
def grnn(x_train, y_train, x_pred, epsilon):
  import pandas as pd
  import numpy as np
  import math

  if isinstance(x_train,pd.DataFrame):
    x_train_num = x_train.to_numpy()
  else:
    x_train_num = x_train

  if isinstance(x_pred,pd.DataFrame):
    x_pred_num = x_pred.to_numpy()
  else:
    x_pred_num = x_pred

  if isinstance(y_train,pd.DataFrame):
    y_train_num = y_train.to_numpy()
  else:
    y_train_num = y_train

  assert x_train_num.dtype == "int64" or x_pred_num.dtype == "int64" or y_train_num == "int64","All data should be in int64 format (you can use: numpy.int64()). Or maybe you have some non-numeric data."

  y_pred = []

  for x_pred_point in x_pred_num:
        euclidean_distance = np.sum((x_train_num-x_pred_point)**2, axis=1)**0.5
        gaussian_kernel = [math.exp(-1*(distance**2)/(2*(epsilon**2))) for distance in euclidean_distance]
        denominator = sum(gaussian_kernel)
        numerator = np.sum(np.array(gaussian_kernel)*y_train_num.flatten())
        y_pred_point = numerator/denominator
        y_pred.append(y_pred_point)

  return np.array(y_pred)
```
